```python
#Import required libraries
import matplotlib.pyplot as plt
%matplot inline
```

```python
#Data from Ohio State -> Leading casues of death in the state from 2012
cause = 'Crhonic Diseases', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia', 'Sepsis','Others'
```

```python
#Percentiles
percentile = [62, 5, 4, 2, 1, 26]
```

```python
#Set the pie chart plot properties
#Set figure size
plt.figure(figsize=(12,12))
#Explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)
#Set pie chart properties
plt.pie(percentile, labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
#Set axis equal to draw pie as circle
plt.axis('equal')
#Set title of the pie chart
plt.title('Ohio State - 2012: Leading Causes of Death', color='white')
#Shoe the plot pie chart
plt.show()
```
